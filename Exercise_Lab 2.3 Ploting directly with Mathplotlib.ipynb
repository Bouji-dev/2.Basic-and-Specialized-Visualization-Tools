```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv"
# resp = await fetch(URL)
# text = io.BytesIO((await resp.arrayBuffer()).to_py())
df_can = pd.read_csv(URL)
```

```python
df_can.head()
```

```python
df_can.shape
```

```python
df_can.set_index('Country', inplace=True)
df_can.head()
```

```python
df_can.shape
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
df_line = df_can[years]
total_immigrants = df_line.sum()
total_immigrants.head()
```

```python
fig , ax = plt.subplots()

ax.plot(total_immigrants)
ax.set_title('Total Immigrants to Canada (1980-2013)')
ax.set_xlabel('Years')
ax.set_ylabel('Number of Immigrants')
```

```python
fig, ax = plt.subplots()

total_immigrants.index = total_immigrants.index.map(int)

ax.plot(total_immigrants)
ax.set_title('Immigrants between 1980 to 2013')

#Setting up the Labels
ax.set_xlabel('Years')
ax.set_ylabel('Total Immigrants')
```

```python
fig, ax = plt.subplots()
total_immigrants.index = total_immigrants.index.map(int)
ax.plot(
    total_immigrants,
    marker='s',
    markersize=5,
    color= 'green',
    linestyle='dotted',
)

ax.set_title('Immigrants between 1980 to 2013')

#Setting up the Labels
ax.set_xlabel('Years')
ax.set_ylabel('Total Immigrants')
ax.legend(['Immigrants'])
```

```python
fig, ax = plt.subplots()
ax.plot(
    total_immigrants,
    marker='s',
    markersize=5,
    color= 'green',
    linestyle='dotted',
)

plt.xlim(1975, 2015)
plt.grid(True)
ax.set_title('Immigrants between 1980 to 2013')

#Setting up the Labels
ax.set_xlabel('Years')
ax.set_ylabel('Total Immigrants')
ax.legend(['Immigrants'])
```

### Question: What is the immigration distribution for China and India for years 2000 to 2013?


```python
y = list(map(str, range(2000, 2013)))
df= df_can.groupby(['Country'])[years].sum()
df_ci = df.loc[['China', 'India'], y]
df_ci = df_ci.T
fig, ax = plt.subplots(figsize= (10, 6))
ax.hist(df_ci)
ax.set_title('Immigration from China and India from 2000 - 2013')
ax.set_xlabel('Number of Immigrants')
ax.set_ylabel('Number of Years')
ax.legend(['China', 'India'])
```

### Question: Create a pie chart representing the total immigrants proportion for each continent

First, you will have to group the data over continents and get the sum on total. Then you can pass this data to the pie function


```python
df_con = df_can.groupby(['Continent'])['Total'].sum().reset_index()
label = list(df_con['Continent'])
label[3] = 'LAC'
label[4] = 'NA'
df_con
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.pie(df_con['Total'], labels=label, autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Immigrants by Continent')
plt.legend(title='Continent', loc='upper right')
```

```python
fig, ax = plt.subplots(1, 2 , figsize=(12, 6))

ax[0].plot(total_immigrants)
ax[0].set_title('Line plot of Total Immigrants')
ax[0].set_ylabel('Number of Immigrants')


ax[1].scatter(total_immigrants.index, total_immigrants)
ax[1].set_title('Scatter plot of Total Immigrants')
fig.suptitle('Subplotting Example', fontsize=15)
fig.tight_layout()

```

```python
fig = plt.figure(figsize=(8,4))

axs1 = fig.add_subplot(1, 2, 1)
axs1.plot(total_immigrants)
axs1.set_title("Line plot on immigrants")

axs2 = fig.add_subplot(1, 2, 2)
axs2.barh(total_immigrants.index, total_immigrants)
axs2.set_title("Bar plot on immigrants")

fig.suptitle('Subplotting Example', fontsize=15)


fig.tight_layout()
```
